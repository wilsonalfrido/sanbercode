```python
#tanpa handling missing value
fig , ax = plt.subplots(figsize=(12,8))
plt.scatter(x = df['HumidityAvgPercent'], y = df['DewPointAvgF'], c = df['TempAvgF'], cmap = 'coolwarm', s = df['WindAvgMPH']*20 )
ax.set(title = 'Austin Weather', xlabel = 'Humidity Average (%)', ylabel = 'Dew Point Average (F)')
plt.show()
```

```python
#dengan handling missing value
df_mv = df[['DewPointAvgF', 'HumidityAvgPercent', 'WindAvgMPH','TempAvgF']].fillna(method='ffill').copy()
fig2 , ax2 = plt.subplots(figsize=(12,8))
plt.scatter(x = df_mv['HumidityAvgPercent'], y = df_mv['DewPointAvgF'], c = df_mv['TempAvgF'], cmap = 'coolwarm', s = df_mv['WindAvgMPH']*20 )
ax2.set(title = 'Austin Weather', xlabel = 'Humidity Average (%)', ylabel = 'Dew Point Average (F)')
plt.show()
```

Dengan menggunakan Interquartile range, kita bisa temukan nilai data outliers dan meremove data outlier tersebut. 

*   Tentukan nilai batas atas dan batas bawah dari Interquartile range tersebut
*   Remove data outliers tersebut menggunakan batas atas dan batas bawah

**EXPECTED OUTPUT:**

Visualisasi boxplot:

![boxplot2](https://drive.google.com/uc?id=1qZc-lEa5FUzSQ_cyt4_GJQ0TZ8c6WjHX)





```python
import seaborn as sns
sns.set(rc={'figure.figsize': (12,8)})
with sns.axes_style('white'):
    ax = sns.boxplot(data = df2['House_Price'], orient = "h").set(xlabel="House_Price")
```

```python
#mencari batas atas dan bawah interquartile
Iqr = df2['House_Price'].describe()['75%'] - df2['House_Price'].describe()['25%']
batas_bawah = df2['House_Price'].describe()['25%'] - (1.5*Iqr)
batas_atas = df2['House_Price'].describe()['75%'] + (1.5*Iqr)
df2_clean = df2[(df2['House_Price'] >= batas_bawah) & (df2['House_Price'] <= batas_atas)].copy()
```

```python
print("Batas atas dan batas bawah:\n\n{0} {1}".format(batas_bawah, batas_atas))
print("\nVisualisasi boxplot:")
sns.set(rc={'figure.figsize': (12,8)})
with sns.axes_style('white'):
    ax2 = sns.boxplot(data = df2_clean['House_Price'], orient = "h").set(xlabel="House_Price")
plt.show()
```
