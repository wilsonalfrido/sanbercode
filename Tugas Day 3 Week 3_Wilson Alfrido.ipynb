```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
fig, ax = plt.subplots(ncols = 3, figsize = (18,6.5), sharey = True)
ax[0].plot(x,tan, color = 'blue', label = 'tanX')
ax[1].plot(x,sin, color = 'red', label = 'sinX')
ax[2].plot(x,cos, color = 'green', label = 'cosX')
ax[0].set_title('Grafik tanx')
ax[1].set_title('Grafik sinx')
ax[2].set_title('Grafik cosx')
fig.legend(loc = 'upper center', bbox_to_anchor=(0.5, 1.03))
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
fig2,ax2 = plt.subplots(figsize = (14,8))
ax2.plot(x,y)
ax2.set_title('Grafik CosX')
ins1 = ax2.inset_axes([-0.1,0.9,0.315,0.28])
ins1.plot(x,y5, color = 'y')
ins1.plot(x,y,color = 'blue')
ins1.set_title('Grafik CosX^5')
ins2 = ax2.inset_axes([0.9,0.9,0.315,0.28])
ins2.plot(x,y4, color = 'black')
ins2.plot(x,y,color = 'blue')
ins2.set_title('Grafik CosX^4')
ins3 = ax2.inset_axes([-0.1,-0.1,0.315,0.28])
ins3.plot(x,y2, color = 'red')
ins3.plot(x,y,color = 'blue')
ins3.set_title('Grafik CosX^2')
ins4 = ax2.inset_axes([0.9,-0.1,0.315,0.28])
ins4.plot(x,y3, color = 'green')
ins4.plot(x,y,color = 'blue')
ins4.set_title('Grafik CosX^3')
plt.show()
```
