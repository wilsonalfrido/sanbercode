Jawaban:

Univariate analysis adalah analisis data hanya pada satu variabel. Karena hanya analisis pada satu variabel saja, maka hasil analisis yang diperoleh bukan bersifat hubungan/sebab-akibat, melainkan untuk mendeskripsikan frekuensi persebaran atau pola dan pada data dalam satu variabel. Hubungan univariate dapat di-visualisasikan menggunakan bar chart, pie chart, polygon, dan histogram

---
Bivariate analysis adalah analisis data yang dilakukan pada 2 variabel data. Tujuan dari analisis tersebut adalah untuk mendeskripsikan hubungan antara 2 variabel. Hubungan variabel tersebut dapat dilihat melalui koefisien korelasi, yang mana jika nilainya mendekati 1, maka terdapat hubungan linear dengan gradien positif , jika -1 maka gradiennya negatif, serta jika mendekati 0, maka tidak terdapat hubungan. Meskipun memiliki hubungan yang kuat, tidak menjadi jaminan bahwa suatu variabel adalah penyebab variabel lainnya terjadi atau biasa disebut "correlation is not causation". Hal tersebut dapat terjadi karena adanya faktor lain tambahan yang sebenarnya mempengaruhi salah satu nilai variabel. Hubungan bivariate dapat digambarkan menggunakan scatter plot, hexbin plot, mosaic plot, heatmaps, dan linechart.


---

Multivariate analysis adalah analisis lebih dari 2 variabel data secara bersamaan. Pada analisis tersebut terdapat minimal satu variabel terikat dan lebih dari satu variabel bebas.

---

Berdasarkan penjelasan di atas, maka terlihat bahwa perbedan ketiga jenis analisis tersebut adalah terletak pada jumlah variabel yang dianalisis.

# Soal 2. Persiapan Data

*   Hapus nilai missing dalam rows dataset menggunakan atribut.dropna()

*   Lakukan grouping data berdasarkan kolom 'species' dengan menggunakan agregasi mean()



```python
#drop missing values
df.dropna(axis = 0, inplace = True)
#grouping data
df_group = df.groupby(['species']).mean().copy()
df_group
```

# Soal 4. Bivariate Analysis



*   Buatlah sebuah grafik catplot menggunakan seaborn dengan x='sex', y='bill_length_mm' dan kind 'bar' seperti expected output di bawah
*   Buatlah sebuah grafik scatter plot menggunakan seaborn dengan x='body_mass_g', y='flipper_length_mm', hue='bill_length_mm' seperti expected dibawah
*   Jelaskan insight yang diperoleh dari hasil analisa bivariate tersebut






```python
print("Catplot")
sns.catplot(data = df, x = 'sex', y = 'bill_length_mm', kind = 'bar')
plt.show()
print("Scatter plot")
plt.figure(figsize = (12,8))
sns.scatterplot(data = df, x='body_mass_g', y='flipper_length_mm', hue = 'bill_length_mm')
plt.show()
```
